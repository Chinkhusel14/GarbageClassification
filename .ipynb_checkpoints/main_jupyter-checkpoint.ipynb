```python
#Imports
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Load and Preprocess
data_dir = './dataset-resized' 
categories = ['cardboard','glass','metal','paper','plastic','trash']

data, labels = [], []

for category in categories:
    path = os.path.join(data_dir, category)
    class_label = categories.index(category)
    for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_COLOR)
            resized_img = cv2.resize(img_array, (128, 128)) 
            data.append(resized_img)
            labels.append(class_label)
        except Exception as e:
            pass

data = np.array(data) / 255.0 
labels = np.array(labels)

X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
#VISUALIZE DATASET
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(data[i])
    plt.title(categories[labels[i]])
    plt.axis('off')
plt.show()
```

```python
#BUILD CNN MODEL
from tensorflow.keras import Input

model = Sequential([
    Input(shape=(128, 128, 3)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(categories), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#Model train
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2, batch_size=32)
```

```python
#TRAIN 2
history = model.fit(X_train, y_train, 
                    epochs=20,
                    validation_split=0.2, 
                    batch_size=32)
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.callbacks import EarlyStopping

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(X_train, y_train, 
                    epochs=50, 
                    validation_split=0.2, 
                    batch_size=32, 
                    callbacks=[early_stopping, reduce_lr])
```

```python
#TEST Preformance
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
def predict_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (128, 128)) / 255.0
    img = np.expand_dims(img, axis=0)
    prediction = model.predict(img)
    return categories[np.argmax(prediction)]

print(predict_image('./dataset-resized/glass/glass1.jpg'))
```

```python
model.save('trash_sorting_model.keras')  # Save the model
```

```python
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```

```python
#Data Augmentation
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

datagen.fit(X_train)
```

```python
#TEST Preformance
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
history = model.fit(datagen.flow(X_train, y_train, batch_size=32), 
                    validation_data=(X_test, y_test),
                    epochs=50)
```

```python
model.save('trash_sorting_model_v2.keras')
```

```python
import random
import matplotlib.pyplot as plt
import os
import cv2

# Function to load random images from dataset
def load_random_images(data_dir, categories, num_images=5):
    images = []
    labels = []
    for category in categories:
        path = os.path.join(data_dir, category)
        img_files = os.listdir(path)
        selected_imgs = random.sample(img_files, num_images)
        for img_file in selected_imgs:
            img_path = os.path.join(path, img_file)
            img_array = cv2.imread(img_path, cv2.IMREAD_COLOR)
            resized_img = cv2.resize(img_array, (128, 128))
            images.append(resized_img)
            labels.append(category)
    return np.array(images), labels

# Load a batch of random images and their labels
num_images = 5  # Number of images to display
random_images, random_labels = load_random_images(data_dir, categories, num_images)

# Normalize images
random_images = random_images / 255.0

# Make predictions
predictions = [predict_image(image_path) for image_path in random_labels]

# Plot images and their predictions
fig, axes = plt.subplots(1, num_images, figsize=(15, 3))
for i in range(num_images):
    ax = axes[i]
    ax.imshow(random_images[i])
    ax.axis('off')
    ax.set_title(f"Predicted: {predictions[i]}")
plt.show()
```
